```python
from fbprophet import Prophet 
```

```python
import pandas as pd
```

```python

dollars= pd.read_csv('dollars.csv')
euros= pd.read_csv('euros.csv')
pounds= pd.read_csv('pounds.csv')

```

```python
dollars.tail()
```

```python
euros.tail()
```

```python
pounds.tail()
```

```python
dollars= dollars[['Date','Close']]
```

```python
dollars['Date']=pd.to_datetime(dollars['Date'])
```

```python
dollars.head()
```

```python
# -------------------------------------------------------dollar------------------------------------------
```

```python
dollar_m=Prophet()
```

```python
dollars.columns=['ds','y']
```

```python
dollar_model=dollar_m.fit(dollars)
```

```python
future_dollar=dollar_m.make_future_dataframe(periods=1825,freq='D')
```

```python
forecast_dollar=dollar_m.predict(future_dollar)
```

```python
forecast_dollar.tail()
```

```python
y_val=forecast_dollar[forecast_dollar['ds']=='2018-10-26']['yhat'].values[0]
```

```python
import plotly.express as px

fig = px.line(forecast_dollar, x="ds", y="yhat", title='dollar forecast over the days ')
#fig.show()

fig.add_annotation(x='2018-10-26', y=y_val,
            text='2018-10-26',
            showarrow=True,
            arrowhead=1)
```

```python
import pickle 

pickle.dump(forecast_dollar, open('dollar_value.pkl','wb'))
```
